```python
# Python script to extract names, locations, and organizations from a NER dataset in CONLL format

input_file = 'Train.txt'

names = set()
locations = set()
organizations = set()

with open(input_file, 'r', encoding='utf-8') as file:
    current_entity = []
    current_type = None
    for line in file:
        line = line.strip()
        if not line:
            if current_entity and current_type:
                entity_text = ' '.join(current_entity)
                if current_type == 'PER':
                    names.add(entity_text)
                elif current_type == 'LOC':
                    locations.add(entity_text)
                elif current_type == 'ORG':
                    organizations.add(entity_text)
            current_entity = []
            current_type = None
            continue
        parts = line.split()
        if len(parts) < 2:
            continue
        word, tag = parts[0], parts[-1]
        if tag.startswith('B-'):
            if current_entity and current_type:
                entity_text = ' '.join(current_entity)
                if current_type == 'PER':
                    names.add(entity_text)
                elif current_type == 'LOC':
                    locations.add(entity_text)
                elif current_type == 'ORG':
                    organizations.add(entity_text)
            current_entity = [word]
            current_type = tag[2:]
        elif tag.startswith('I-') and current_type == tag[2:]:
            current_entity.append(word)
        else:
            if current_entity and current_type:
                entity_text = ' '.join(current_entity)
                if current_type == 'PER':
                    names.add(entity_text)
                elif current_type == 'LOC':
                    locations.add(entity_text)
                elif current_type == 'ORG':
                    organizations.add(entity_text)
            current_entity = []
            current_type = None

with open('names.txt', 'w', encoding='utf-8') as f:
    for name in sorted(names):
        f.write(name + '\n')

with open('locations.txt', 'w', encoding='utf-8') as f:
    for loc in sorted(locations):
        f.write(loc + '\n')

with open('organizations.txt', 'w', encoding='utf-8') as f:
    for org in sorted(organizations):
        f.write(org + '\n')
```
